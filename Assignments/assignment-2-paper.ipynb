### 3.1. cbind()

```r
c1 <- c(1, 8, 3)
c2 <- c(2, 7, 3)
c3 <- c(3, 6, 3)
c4 <- c(4, 5, 3)
c5 <- c(5, 4, 3)
c6 <- c(6, 3, 3)
c7 <- c(7, 2, 3)
c8 <- c(8, 1, 3)
cbind_matrix <- matrix(cbind(c1, c2, c3, c4, c5, c6, c7, c8), 3)
cbind_matrix
```

```r
r1 <- c(1, 2, 3, 4, 5, 6, 7, 8)
r2 <- c(8, 7, 6, 5, 4, 3, 2, 1)
r3 <- c(3, 3, 3, 3, 3, 3, 3, 3)
rbind_matrix <- matrix(rbind(r1, r2, r3), 3)
rbind_matrix
```

## 11. Program to get the weight and height from the user

```r
calcBMI <- function(weight, height) {
  return(round(weight / (height * height), 0))
}

weight <- readline(prompt = "Enter your weight (kg): ")
height <- readline(prompt = "Enter your height (m): ")

weight <- as.double(weight)
height <- as.double(height)

bmi = calcBMI(weight = weight, height = height)

if (bmi < 18.5) {
    print(paste('BMI: ', bmi, ' - under weight'))
} else if (18.5 < bmi && bmi < 25) {
    print(paste('BMI: ', bmi, ' - normal'))
} else if (25 < bmi && bmi < 30) {
    print(paste('BMI: ', bmi, ' - over weight'))
} else if (30 < bmi) {
    print(paste('BMI: ', bmi, ' - obese'))
}
```

## 13. While loop vs repeat

### 13.1 while loop

```r
# In R programming, while loops are used to loop until a specific condition is met.
i <- 0
while (i!=2) {print(i);i<-i+1}
```

### 13.2 repeat

```r
# A repeat loop is used to iterate over a block of code multiple number of times.

# There is no condition check in repeat loop to exit the loop.

# We must ourselves put a condition explicitly inside the body of the loop and use the break statement to exit the loop

j <- 1
repeat {
    print(j)
    j <- j+1
    if (j == 6){
        break
    }
}
```
